```python
import boto3
from botocore import UNSIGNED
from botocore.config import Config
```

```python
# Configure S3 client for unsigned requests (public buckets)
s3 = boto3.client('s3', 
                  config=Config(signature_version=UNSIGNED))
```

```python

def scan_public_bucket(bucket_name, prefix=''):
    """Scan public S3 bucket without authentication"""
    files = []
    
    try:
        paginator = s3.get_paginator('list_objects_v2')
        pages = paginator.paginate(Bucket=bucket_name, Prefix=prefix)
        
        for page in pages:
            if 'Contents' in page:
                for obj in page['Contents']:
                    files.append({
                        'key': obj['Key'],
                        'size': obj['Size'],
                        'last_modified': obj['LastModified'],
                        'full_path': f"s3://{bucket_name}/{obj['Key']}"
                    })
        
        return files
        
    except Exception as e:
        print(f"Error scanning bucket: {e}")
        return []

# Usage - scan the Snowflake quickstart bucket
# bucket_name = 'sfquickstarts'
bucket_name = "tripdata"
prefix = ''

files = scan_public_bucket(bucket_name, prefix)

print(f"Found {len(files)} files:")
for file in files:
    print(f"  {file['full_path']} ({file['size']} bytes)")
```
